```python
!pip install matminer pandas scikit-learn
import pandas as pd
from matminer.featurizers.conversions import StrToComposition
from matminer.featurizers.composition import ElementProperty

df = pd.read_csv("mp_hardness_data.csv")

# 2. Convert 'formula' strings to Pymatgen Composition objects
# Instead of .execute(), we use .featurize_dataframe()
stc = StrToComposition()
df = stc.featurize_dataframe(df, "formula") 

# 3. Generate the Magpie features from the 'composition' column
ep_feat = ElementProperty(
    data_source="magpie",
    features=["AtomicWeight", "Number", "Electronegativity", "MendeleevNumber", "MeltingT"],
    stats=["mean", "avg_dev"]
)
df = ep_feat.featurize_dataframe(df, col_id="composition")

# Check your columns to see the new features
print(df.head())
df.to_csv("mp_hardness_with_features.csv", index=False)
```
